```python
from typing import Final
from telegram import Update, InputFile, BotCommand
from telegram.ext import Application, CommandHandler, MessageHandler, filters, ContextTypes, ConversationHandler
import nest_asyncio

# necessary libraries for extra functionality
import numpy as np
import matplotlib.pyplot as plt
import io
from time import sleep

###################################################

# bot token and username
TOKEN: Final = {TOKEN}
BOT_USERNAME: Final = {BOT_USERNAME}

# conversation states
ASKING_FOR_ARRAY = 0
ASKING_FOR_SIZE = 1
ASKING_FOR_CMAP = 2

####################################################################

# defining the main bot commands
async def start_command(update: Update, context: ContextTypes.DEFAULT_TYPE):
    await update.message.reply_text("bQuasar's here!! The stars have aligned for you today *** ")

async def help_command(update: Update, context: ContextTypes.DEFAULT_TYPE):
    await update.message.reply_text("space themed/\nb stands for bot, profile picture's generated with Adobe Firefly, functionality's in\n/info")

async def info_command(update: Update, context: ContextTypes.DEFAULT_TYPE):
    await update.message.reply_text("I have the universal power to create images from numpy arrays using the matplotlib.pyplot library ((()))")

# defining the custom command
async def create_command(update: Update, context: ContextTypes.DEFAULT_TYPE):
    # clear previous data
    context.user_data.clear()
    await update.message.reply_text("Provide me a numpy matrix [[*],[*]]")
    return ASKING_FOR_ARRAY

async def handle_array_input(update: Update, context: ContextTypes.DEFAULT_TYPE):
    try:
        # extracting the numpy matrix from the message text
        array_text = update.message.text
        context.user_data['array'] = np.array(eval(array_text))

        # asking for image size
        await update.message.reply_text("Provide me an image size *,*")
        return ASKING_FOR_SIZE

    except Exception as e:
        await update.message.reply_text(f"An error occurred: {str(e)}")
        return ConversationHandler.END

async def handle_size_input(update: Update, context: ContextTypes.DEFAULT_TYPE):
    try:
        # extracting image size from the message text
        size_text = update.message.text
        width, height = map(int, size_text.split(','))
        context.user_data['size'] = (width, height)

        # asking for colormap
        await update.message.reply_text("Provide me a color map viridis, gray, *")
        return ASKING_FOR_CMAP

    except Exception as e:
        await update.message.reply_text(f"An error occurred: {str(e)}")
        return ConversationHandler.END

async def handle_cmap_input(update: Update, context: ContextTypes.DEFAULT_TYPE):
    try:
        # extracting colormap from the message text
        cmap = update.message.text
        array = context.user_data['array']
        size = context.user_data['size']

        # creating an image
        fig, ax = plt.subplots(figsize=size)
        ax.imshow(array, cmap=cmap)
        plt.axis('off')

        # saving the image to a BytesIO object
        image_stream = io.BytesIO()
        plt.savefig(image_stream, format='png', bbox_inches='tight', pad_inches=0)
        image_stream.seek(0)
        plt.close(fig)

        # sending the image back to the user
        await update.message.reply_photo(photo=InputFile(image_stream, filename="image.png"))

        # asking for colormap again for infinite image creation
        await update.message.reply_text("milky way ain't the only galaxy in our space.  .    .")
        return ASKING_FOR_CMAP

    except Exception as e:
        await update.message.reply_text(f"An error occurred: {str(e)}")
        return ConversationHandler.END

######################################################

# defining bot responses private chat
def handle_response(text: str) -> str:
    processed: str = text.lower()

    if "hello" in processed:
        return "Hi i   i     i       i"

    if "how are you" in processed:
        return "Stellar! How about you?"

    if "bye" in processed:
        return "Signal lost. you fell into a black hole {}"

    return "Low gravity: couldn't catch your response."

# defining bot responses group chat
async def handle_message(update: Update, context: ContextTypes.DEFAULT_TYPE):
    message_type: str = update.message.chat.type
    text: str = update.message.text

    if message_type == 'group':
        # check if the bot's username is mentioned in the message
        if BOT_USERNAME.lower() in text.lower():
            new_text: str = text.replace(BOT_USERNAME, '').strip()
            response: str = handle_response(new_text)
            # mention the user who sent the original message
            user_mention = f"@{update.message.from_user.username}"
            await update.message.reply_text(f"{user_mention} {response}")
    else:
        response: str = handle_response(text)
        await update.message.reply_text(response)

########################################################

# handling errors so the bot doesn't crash
async def error(update: Update, context: ContextTypes.DEFAULT_TYPE):
    print(f"Update caused error: {context.error}")

#########################################

if __name__ == '__main__':
    sleep(0.5)
    print("Starting the bot.....")
    app = Application.builder().token(TOKEN).build()

    # conversation handler for the create command
    create_conv_handler = ConversationHandler(
        entry_points=[CommandHandler('create', create_command)],
        states={
            ASKING_FOR_ARRAY: [MessageHandler(filters.TEXT & ~filters.COMMAND, handle_array_input)],
            ASKING_FOR_SIZE: [MessageHandler(filters.TEXT & ~filters.COMMAND, handle_size_input)],
            ASKING_FOR_CMAP: [MessageHandler(filters.TEXT & ~filters.COMMAND, handle_cmap_input)],
        },
        fallbacks=[]
    )

    # commands
    app.add_handler(CommandHandler('start', start_command))
    app.add_handler(CommandHandler('help', help_command))
    app.add_handler(CommandHandler('info', info_command))
    app.add_handler(create_conv_handler)

    # messages
    app.add_handler(MessageHandler(filters.TEXT, handle_message))

    # errors
    app.add_error_handler(error)

    # polling the bot
    sleep(1.5)
    print("Polling....")
    nest_asyncio.apply()
    sleep(1)
    print("Ready")

    app.run_polling()
```
